```python
%matplotlib inline

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('data/sheep-vary-n-l-full.csv', skiprows=6)
```

```python
plt.imshow(
    data[data['[step]'] > 500].groupby(['sheep-sim-n', 'sheep-sim-l'])[['count sheep', 'count wolves', 'count grass']]
    .mean()
    .reset_index()
    .pivot_table(
        values='count sheep',
        columns='sheep-sim-l',
        index='sheep-sim-n',
    ),
#  .plot(
#     #y='count sheep',
#     #xticks = data['sheep-sim-n'].unique(),
#     #subplots=True,
#     #figsize=(8,14)
# )
    extent=(data['sheep-sim-l'].min() - 0.5, data['sheep-sim-l'].max() + 0.5, data['sheep-sim-n'].min() - 0.5, data['sheep-sim-n'].max() + 0.5),
    origin = 'lower',
)

plt.xticks(data['sheep-sim-l'].unique())
plt.yticks(data['sheep-sim-n'].unique())
plt.xlabel('micro-sim-length')
plt.ylabel('num-micro-sims')
plt.colorbar()
```
